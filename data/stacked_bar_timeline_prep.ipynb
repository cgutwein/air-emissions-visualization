```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
df = pd.read_csv("GEN14_3.csv")
```

```python
df.head()
```

```python
## 'NAMEPCAP' -- Nameplate capacity
## 'GENYRRET' -- Retirement Year
## 'GENYRONL' -- Year Unit went online
## 'FUELG1' -- Primary Fuel Type
## 
## We want a new dataframe with
## one row for each year
## and a column for each fuel type
f_types = df['FUELG1'].unique()
years = df['GENYRONL'].unique()
years = sorted(years)
print(years)
#print(f_types)
first_year = 1891.
print(first_year)
max_year = 2030.
```

```python
years = np.array(list(range(int(first_year), int(max_year + 1))))
df_new = pd.DataFrame(np.zeros((len(years), len(f_types))), index = years)
df_new.columns = f_types
print(df_new.head())
```

```python
print(df.iloc[1]['GENYRONL'])
print(df_new.loc[[1891.0]]['WC'])
print(type(df.iloc[1]['NAMEPCAP']))
## preprocess values in NAMEPCAP
```

```python
for i in range(len(df)):
    if df.iloc[i]['GENYRONL'] in years:
        val = float(re.sub(',*','', df.iloc[i]['NAMEPCAP']))
        print(val)
        yr = df.iloc[i]['GENYRONL']
        fuel = df.iloc[i]['FUELG1']
        df_new.loc[yr][fuel] += val
    else:
        continue
```

```python
df_new['Coal'] = df_new['BIT'] + df_new['SUB'] + df_new['LIG'] + df_new['RC']+df_new['WC']
df_new['Nat_gas'] = df_new['NG'] + df_new['PG'] + df_new['SGC'] + df_new['SGP']
df_new['Petroleum'] = df_new['DFO'] + df_new['JF'] + df_new['KER'] + df_new['PC'] + df_new['RFO'] + df_new['WO']
df_new['Water'] = df_new['WAT']
df_new['Solar'] = df_new['SUN']
df_new['Wind'] = df_new['WND']
df_new['Nuclear'] = df_new['NUC']
df_new['Other'] = df_new['AB'] + df_new['BFG'] + df_new['BLQ'] + df_new['GEO'] + df_new['LFG'] + df_new['MSW'] + df_new['MWH'] + df_new['OBG'] + df_new['OBL'] + df_new['OBS'] + df_new['OG'] + df_new['OTH'] + df_new['PUR'] + df_new['SLW'] + df_new['TDF'] + df_new['WDL'] + df_new['WDS'] + df_new['WH']
```

```python
df_new_1 = df_new.drop(f_types, axis=1)
df_new_1.columns
```
